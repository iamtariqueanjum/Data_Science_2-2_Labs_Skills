```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv')
df.head()
```

**1.Information/summary**

```python
df.info()
df.describe()
```

**2.Mode of Health and beauty, accessories, Electronic , Home and lifestyle, Food and beverages, Sports**

```python
from scipy import stats
mode =stats.mode(df['Product line'])
mode
```

**3.Mean of Health and beauty, accessories, Electronic , Home and lifestyle, Food and beverages, Sports**

```python
mean = df.groupby('Product line').agg([np.mean])
mean
```

**4.Standard deviation of unit price**

```python
sd = df['Unit price'].std()
print("Standard Deviation of unit price: ",sd)
```
