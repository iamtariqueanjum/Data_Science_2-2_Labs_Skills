**6.Find ‘Timeliness of care national comparison footnote’ and its ‘hospital’**

```python
time = pd.crosstab(hospital_ratings['Hospital Name'],hospital_ratings['Timeliness of care national comparison footnote']=='Above the national average')
time
```

**7.‘Hospital rating’**

**8.Find ‘Acute Care Hospitals’ with rating (4) along with ’ state’**

```python
hospital_ratings.loc[(hospital_ratings['Hospital Type']=='Acute Care Hospitals')&(hospital_ratings['Hospital overall rating']=='4'),['Hospital Name','State','Hospital overall rating']]
```

**9.Which type of hospitals has highest ‘Non-availability’ of ‘Safety of Care data’?**

```python
Non_available = hospital_ratings.groupby('Hospital Type')['Safety of care national comparison'].count()
Non_available.sort_values(ascending=False)
```
