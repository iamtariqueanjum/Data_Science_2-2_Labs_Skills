```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("NHANES.csv")
```

```python
data.head()
```

```python
for i in data.select_dtypes(include='int' ):
    sns. distplot(data[i])
    plt. show()
```

```python
for i in data. select_dtypes(include='int') :
    sns.boxplot(data [i])
    plt.show()
```

```python
for i in data.select_dtypes(include='float'):
    sns.distplot(data[i])
    plt.show()
```

```python
data.hist()
```

```python
plt.hexbin(data["SEQN"], data["BMXWT"], gridsize=(25,25), cmap=plt.cm.Greens)
plt.show()
```

```python
d1 = data["SEQN"]   
d2 = data["BMXWT"]                      
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]    

plt.pie(d1 , labels=d2 , colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.show()
```
