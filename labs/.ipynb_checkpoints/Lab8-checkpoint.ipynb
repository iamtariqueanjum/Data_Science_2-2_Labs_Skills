```python
import numpy as np
import pandas as pd
```

```python
df=iris.loc[iris['species']=='Iris-setosa']
df
```

```python
df.species.unique()
```

```python
unique_sepal_lengths=df.sepal_length.unique()
unique_sepal_lengths
```

```python
frequency=pd.crosstab(index=df['sepal_length'],columns=["count"])
frequency
```

```python
total=df['species'].count()
total
```

```python
probabilities = frequency['count']/total
probabilities
```

```python
fre = []
pro = []
c = 0
count = np.count_nonzero(df.sepal_length)
a = df.sepal_length.unique()
for i in a:
    for j in df['sepal_length']:
        if i == j:
            c = c + 1
    fre.append(c)
    pro.append(c/count)
print(fre)
print(pro)
df1 = pd.DataFrame({'sepal_length':a,'frequency':fre,'probability':pro})
df1.head()
```

```python
expect_value = 0
i = 0
for val in a:
    expect_value = expect_value + (val*pro[i])
    i = i + 1
print(expect_value)    
```

```python
df2 = iris.loc[iris['species'] == 'Iris-virginica']
count1 = np.count_nonzero(df2.petal_length)
b = df2.petal_length.unique()
b
```

```python
fre1 = []
pro1 = []
c1 = 0
for i in b:
    for j in df2['petal_length']:
        if i == j :
            c1 = c1 + 1
    fre1.append(c1)
    pro1.append(c1/count1)
print(fre1)
print(pro1)
```

```python
expect_value1 = 0
i = 0
for val in b:
    expect_value1 = expect_value1 + (val * pro1[i])
    i = i + 1
print(expect_value1)
```
